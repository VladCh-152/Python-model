```python
class hypothesis(object):
    def __init__(self):
        self.b0 = 0
        self.b1 = 1
    def predict(self, x):
        return self.b0 + self.b1 * x
    def error(self, X, Y):
        return np.sum((self.predict(X) - Y)**2) / (2 * len(X))
    def BGD(self, X, Y, alpha=0.5, accuracy=0.1, max_steps=10000):
        step = 0
        old_err = self.error(X, Y)
        new_err = self.error(X, Y)
        dJ = 1
        while (dJ > accuracy) and (step < max_steps):
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1
            old_err = new_err
            new_err = self.error(X, Y)
            dJ = abs(old_err - new_err) 
            step += 1
```

```python
def create_series(data, last, num):
    series = np.zeros(num)
    for i in range(num):
        series[i] = data['Close'][len(data['Close'])-num+i]
    return series
```

```python
mx = 0
for i in range(1, len(normalized_h)):
    if normalized_h[i]-normalized_h[i-1]>mx:
        mx = np.abs(normalized_h[i]-normalized_h[i-1])
print(mx)
```

```python
koef = -256
def create_dataset(data, num):
    dataset = np.ones((len(data)-num-4, num))
    targets = np.zeros((len(data)-num-4, 4))
    for i in range(len(dataset)-1):
        dataset[i][0] = data[i]
        for j in range(1, num):
            dataset[i][j] = 1/(1+np.exp(koef*(data[i+j]-data[i+j-1])))
        for j in range(4):
            targets[i][j] = 1/(1+np.exp(koef*(data[i+num+j]-data[i+num-1])))
    return dataset, targets
```

```python
model_inputs, targets = create_dataset(normalized_h, 12)
        
train_inputs = model_inputs[:-10]
val_inputs = model_inputs[-10:]
train_targets = targets[:-10]
val_targets = targets[-10:]

train_targets-0.5
```

```python
def create_model(n):
    m = keras.Sequential()
    m.add(layers.Dense(64, activation='relu', input_shape=(n,)))
    #m.add(Dropout(0.1))
    m.add(layers.Dense(128, activation='relu'))
    #m.add(Dropout(0.1))
    m.add(layers.Dense(4))
    return m
```

```python
model_inputs, targets = create_dataset(normalized_h, 5)
print(np.sum(targets)/len(targets))
```

```python
k = 200
p = 9000
s1, f1 = 4, 144
epochs = 100
def select_inputs(data, start, end, epochs):
    models = {}
    for inputs in range(start, end+1):
        print('Using {} inputs'.format(inputs))
        model_inputs, targets = create_dataset(data, inputs)
        
        train_inputs = np.concatenate((model_inputs[0:p], model_inputs[p+k:-1]))
        val_inputs = model_inputs[p:p+k]
        train_targets = np.concatenate((targets[0:p], targets[p+k:-1]))
        val_targets = targets[p:p+k]
        
        m = create_model(inputs)
        print('Training')
        m.compile(optimizer='adam', loss='mse', metrics=["accuracy"])
        h = m.fit(train_inputs, train_targets,
                  epochs=epochs,
                  batch_size=32,
                  validation_data=(val_inputs, val_targets))
        model_info = {'model': m, 'history': h.history}
        models[inputs] = model_info
    return models
```

```python
trained_models = select_inputs(normalized_h, s1, f1, epochs)
```

```python
for i in range(s1, f1+1):
    name = "model_" + str(i) + ".h5"
    trained_models[i]['model'].save(name)
```

```python
model_stats = {}
ac = []
for k, v in trained_models.items():
    train_history = v['history']
    loss = train_history['loss'][-1]
    acc = train_history['val_accuracy'][-1]
    ac.append([acc, k])
    val_loss = train_history['val_loss'][-1]
    model_stats[k] = {'inputs': k, 'loss': loss, 'accuracy': acc, 'val_loss': val_loss}
```
